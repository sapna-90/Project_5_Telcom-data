```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from scipy.spatial.distance import euclidean
```

```python
Dataset_4 = pd.read_csv(r"/content/Project_5 - Data_cleaned.csv")
```

```python
Dataset_4.head(5)
```

```python
# Assuming the data contains relevant features for clustering
# Select relevant columns for engagement and experience
engagement_columns = ['IMSI', 'MSISDN/Number', 'Total UL (Bytes)']
experience_columns = ['IMSI', 'MSISDN/Number', 'Avg RTT DL (ms)']
```

```python
engagement_data = Dataset_4[engagement_columns]
experience_data = Dataset_4[experience_columns]
```

```python
# Step 1: Perform K-means clustering to identify the less engaged and worst experience clusters
kmeans_engagement = KMeans(n_clusters=2, random_state=0).fit(engagement_data)
kmeans_experience = KMeans(n_clusters=2, random_state=0).fit(experience_data)
```

```python
# Step 2: Find the centroids of the clusters
engagement_centroids = kmeans_engagement.cluster_centers_
experience_centroids = kmeans_experience.cluster_centers_
```

```python
# Assume the first cluster represents less engaged and worst experience clusters
less_engaged_centroid = engagement_centroids[0]
worst_experience_centroid = experience_centroids[0]
```

```python
# Step 3: Calculate engagement and experience scores (Euclidean distance from the centroids)
Dataset_4['engagement_score'] = engagement_data.apply(lambda row: euclidean(row, less_engaged_centroid), axis=1)
Dataset_4['experience_score'] = experience_data.apply(lambda row: euclidean(row, worst_experience_centroid), axis=1)
```

```python
Dataset_4.head()
```

```python
# Calculate Satisfaction Score and Report Top 10 Users

# Step 4: Compute the satisfaction score as the average of engagement and experience scores
Dataset_4['satisfaction_score'] = (Dataset_4['engagement_score'] + Dataset_4['experience_score']) / 2
```

```python
# Step 5: Sort users by satisfaction score and get the top 10 satisfied users
top_10_satisfied_users = Dataset_4.sort_values(by='satisfaction_score', ascending=False).head(10)
```

```python
top_10_satisfied_users[['Bearer Id', 'engagement_score', 'experience_score', 'satisfaction_score']]
```

```python
# Building a regression model to predict satisfaction score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
# Split the data into training and test sets
X = Dataset_4[['engagement_score', 'experience_score']]  # Features
y = Dataset_4['satisfaction_score']  # Target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Train a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict on test data
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
print(f" R2 Score {r2_score(predictions, y_test)}")
print(f" mean_squared_error {mean_squared_error(predictions, y_test)}")
print(f" Root mean_squared_error {np.sqrt(mean_squared_error(predictions, y_test))}")
print(f" mean_absolute_error {mean_absolute_error(predictions, y_test)}")
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Random Forest Regressor model
rf_model = RandomForestRegressor(n_estimators=100, random_state=20)
rf_model.fit(X_train, y_train)
```

```python
# Predicting on the test set
predictions_rf = rf_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
print(f" R2 Score {r2_score(predictions_rf, y_test)}")
print(f" mean_squared_error {mean_squared_error(predictions_rf, y_test)}")
print(f" Root mean_squared_error {np.sqrt(mean_squared_error(predictions_rf, y_test))}")
print(f" mean_absolute_error {mean_absolute_error(predictions_rf, y_test)}")
```

```python
# K-means Clustering on Engagement and Experience Scores

# Perform K-means clustering with k=2
kmeans_satisfaction = KMeans(n_clusters=2, random_state=0).fit(Dataset_4[['engagement_score', 'experience_score']])
```

```python
# Assign cluster labels to users
Dataset_4['cluster'] = kmeans_satisfaction.labels_
```

```python
# Group by cluster and calculate the average engagement, experience, and satisfaction scores
cluster_averages = Dataset_4.groupby('cluster').agg({
    'engagement_score': 'mean',
    'experience_score': 'mean',
    'satisfaction_score': 'mean'
})

cluster_averages
```
